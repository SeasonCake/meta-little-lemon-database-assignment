```python
import mysql.connector as connector

connection = connector.connect(
    host='127.0.0.1',
    port=3306,
    user='root',
    password='YOUR_PASSWORD',
    database='little_lemon'
)
cursor = connection.cursor()
print("Connected to:", connection.database)
```

```python
cursor.execute("SHOW TABLES")
for table in cursor.fetchall():
    print(table)
```

```python
cursor.execute("SHOW PROCEDURE STATUS WHERE Db = 'little_lemon'")
columns = cursor.column_names
for row in cursor.fetchall():
    print(row[1]) 
```

```python
cursor.execute("CALL GetMaxQuantity()")
result = cursor.fetchone()
print(result)
while cursor.nextset():
    pass
```

```python
query = """
SELECT c.CustomerID, c.FullName, o.OrderID, o.Cost
FROM customers c
INNER JOIN orders o ON c.CustomerID = o.CustomerID
WHERE o.Cost > 150
ORDER BY o.Cost ASC;
"""
cursor.execute(query)
for row in cursor.fetchmany(50):
    print(row)
while cursor.nextset():
    pass
```

```python
cursor.execute("DROP PROCEDURE IF EXISTS AddBooking")
while cursor.nextset():
    pass

cursor.execute("""
CREATE PROCEDURE AddBooking(IN booking_id INT, IN customer_id VARCHAR(20), IN table_number INT, IN booking_date DATE)
BEGIN
    INSERT INTO Bookings (BookingID, BookingDate, TableNumber, CustomerID, StaffID)
    VALUES (booking_id, booking_date, table_number, customer_id, 1);
    SELECT 'New booking added' AS Confirmation;
END
""")
while cursor.nextset():
    pass

cursor.execute("SELECT CustomerID FROM Customers LIMIT 1")
sample_cid = cursor.fetchone()[0]

add_booking_query = "CALL AddBooking(%s, %s, %s, %s)"
cursor.execute(add_booking_query, (99, sample_cid, 99, '2022-12-10'))
result = cursor.fetchall()
print(result)
while cursor.nextset():
    pass
```

```python
update_booking_query = "CALL UpdateBooking(%s, %s)"
cursor.execute(update_booking_query, (99, '2022-01-10'))
result = cursor.fetchall()
print(result)
while cursor.nextset():
    pass
```

```python
cancel_booking_query = "CALL CancelBooking(%s)"
cursor.execute(cancel_booking_query, (99,))
result = cursor.fetchall()
print(result)
while cursor.nextset():
    pass
```
